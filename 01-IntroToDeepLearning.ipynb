```python
import tensorflow as tf
from tensorflow import keras
from matplotlib import pyplot as plt
```

```python
x1 = tf.linspace(-10.0, 10.0, 100)
x2 = tf.linspace(-20.0, 20.0, 100)
y = tf.linspace(5, 10, 100)
# random data-> linspace is used to generate random distributed data between 2 limits
```

```python
len(x1), len(x2), len(y)
```

```python
x1[:5]
```

```python
model = keras.Sequential([
    keras.layers.InputLayer(input_shape=(2,)),
    keras.layers.Dense(8, activation='relu'),
    keras.layers.Dense(1)
])

x = tf.stack([x1, x2], axis=1)
model.compile(optimizer='adam', loss='mse')
history = model.fit(x, y, epochs=1000)
```

```python
# uncomment below if you want to see the loss' data structure
#print(history.history)
loss = history.history['loss']
```

```python
epochs = range(1, 1001)

plt.figure(figsize=(10, 6))
plt.plot(epochs, loss, label='Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss vs Epochs')
plt.legend()
plt.grid(True)
plt.show()
```

```python

# Define class names
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle Boots']

# Create a dictionary to map labels to class names
label_to_class = {i: class_name for i, class_name in enumerate(class_names)}

# Display the class names
label_to_class
```

```python
# validate that indeed the remaining 10,000 element pairs represent testing data
print(label_to_class[data[1][1][1]])
plt.imshow(data[1][0][1])
```

```python
# input layer is always the size of input
print(data[0][0][0].shape)
# output layer is always the number of classes
print(len(class_names))
# in between it's hidden layer which is up to us!

# DENSE LAYERS-> accept only 1-D inputs!
# but are images 1D or 2D? = ? = 2D

# FIND A WAY to convert this MATRIX into 1-D data!

# FLATTEN LAYER -> reduces matrices into 1-D flat lines!

# [ 1, 2, 3]
# [ 4, 5, 6].  ==>.  [ 1, 2, 3, 4,5 ,6,....9]
# [ 7, 8, 9]
```

```python
input_layer =  keras.layers.Flatten(input_shape=(28, 28))

hidden_1 = keras.layers.Dense(8, activation='relu')
hidden_2 = keras.layers.Dense(128, activation='relu')

output_layer = keras.layers.Dense(10, activation='softmax')


# ReLU -> remove negative numbers 
# SoftMax -> divide everything into a probability distribution -> sum of all values will be 1
# SIGMOID -> convert everything into an exponential probability function -> 0 to 1
# TanH -> hyperbolic tangent function-> converts everything into a exponential distribution-> - 1 to 1

model = keras.Sequential([
    input_layer, hidden_1, hidden_2, output_layer
])
```

```python
model.compile(loss= 'sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
xtrain = data[0][0]
ytrain = data[0][1]
history = model.fit(xtrain, ytrain, epochs=20, verbose=0) # verbose 0 means don't print out, just calculate!
```

```python
print(history.history)
```

```python
loss = history.history['loss']
accuracy = history.history['accuracy']
```

```python

epochs = range(1, 21)

plt.figure(figsize=(10, 5))
plt.plot(epochs, loss, 'r', label='Loss')
plt.plot(epochs, accuracy, 'b', label='Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Value')
plt.title('Loss and Accuracy over 20 Epochs')
plt.legend()
plt.grid(True)
plt.show()
```
